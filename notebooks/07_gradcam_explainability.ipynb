# Step 9: Explainability with Grad-CAM

Apply Grad-CAM to the final ResNet1D conv layer using PyTorch Captum.

Clinical verification:
- Supraventricular (S): heatmap peaks at P-wave region
- Ventricular (V): heatmap peaks at QRS complex region

One correct + one misclassified example per class (10 plots total).

```python
import sys, os
from pathlib import Path
_r = Path.cwd()
while not (_r / 'src').is_dir():
    _r = _r.parent
os.chdir(_r)
if str(_r) not in sys.path:
    sys.path.insert(0, str(_r))

from src.explainability.examples import get_class_examples
from src.utils.io import load_artifact
from src.utils.device import get_device
import numpy as np
```

```python
from src.data.loader import load_mitbih
from src.data.preprocessor import preprocess
from src.data.dataset import build_dataloaders

X_train_raw, y_train_raw, X_test_raw, y_test_raw = load_mitbih()
splits = preprocess(X_train_raw, y_train_raw, X_test_raw, y_test_raw, val_fraction=0.15)
train_loader, val_loader, test_loader = build_dataloaders(splits, batch_size=128)
print(f'Data ready — test: {splits["X_test"].shape}')
```

```python
from src.models.resnet1d import ResNet1D
from src.evaluation.metrics import get_predictions

device = get_device()
model = ResNet1D()
model.load_state_dict(load_artifact('results/checkpoints/resnet1d_adam_best.pth'))
model = model.to(device)
model.eval()
y_true, y_pred, y_prob = get_predictions(model, test_loader, device)

summary = get_class_examples(
    model, splits['X_test'], y_true, y_pred, y_prob, device,
    save_dir='results/plots/gradcam'
)
print('Grad-CAM examples generated:', summary)
```
